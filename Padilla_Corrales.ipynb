```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

def crear_img1():
  # Tamaño de la imagen
  ancho = 30
  alto = 20

  # Crear una imagen en blanco
  imagen = Image.new('RGB', (ancho, alto), 'white')

  # Dibujar la diagonal en negro
  dibujo = ImageDraw.Draw(imagen)
  flag = True
  a = 0
  for i in range(alto):
    if flag:
      dibujo.line([(a, i), (ancho, i)], fill="black")
      a = a+1
      flag = False
    else:
      dibujo.line([(a, i), (ancho, i)], fill="black")
      a = a+2
      flag = True
  return imagen
imagen = crear_img1()

plt.imshow(imagen)
```

```python
def crear_img2():
  ancho = 30
  alto = 20
  imagen2 = Image.new('RGB', (ancho, alto), 'white')
  # Dibujar la diagonal en negro
  dibujo2 = ImageDraw.Draw(imagen2)
  flag = True
  a = 0
  color = (0,0,0)
  for i in range(20):
    if i > 10:
      color = (85,85,85)
      dibujo2.line([(a, i), (ancho, i)], fill=(170,170,170))
    else:
      dibujo2.line([(29-a, i), (ancho, i)], fill=(170,170,170))

    if flag:
      dibujo2.line([(a, i), (29-a, i)], fill=color)
      a += 1
      flag = False
    else:
      dibujo2.line([(a, i), (29-a, i)], fill=color)
      a += 2
      flag = True
  return imagen2
imagen2 = crear_img2()
plt.imshow(imagen2)
```

```python
def crear_img3():
  # Crear una imagen de 20x11 con valores de intensidad de 0 a 1 en incrementos de 0.1
  imagen3 = Image.new("L", (20, 11))
  for x in range(20):
      for y in range(11):
          valor = int(y * 25.5)  # Convertir 0-10 a 0-255 en incrementos de 25.5
          imagen3.putpixel((x, y), valor)
  return imagen3
imagen3 = crear_img3()
plt.imshow(imagen3)
```

```python
# Crear una imagen con valores aleatorios entre 0 y 1
import random
def crear_img4():
  imagen4 = Image.new("L", (30, 20))
  for y in range(20):
      for x in range(30):
          valor = int(random.uniform(0, 1)*2)
          imagen4.putpixel((x, y), valor * 255)
  return imagen4
imagen4 = crear_img4()
plt.imshow(imagen4)
```

```python
import cv2
def escala_gris(imagen):
  imagen=cv2.imread(imagen)
  imagen_gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
  plt.imshow(imagen_gris,cmap="gray")

escala_gris("mario.jpg")
escala_gris("globo.jpg")
```

```python
#Histograma
import cv2
import matplotlib.pyplot as plt
import numpy as np

def histograma(imagen, canal):
    # Lee la imagen
    img = cv2.imread(imagen)

    # Extrae el canal especificado
    canal_img = img[:, :, canal]

    # Calcula el histograma
    hist = cv2.calcHist([canal_img], [0], None, [256], [0, 256])

    # Convierte el histograma a un arreglo NumPy
    hist = np.array(hist)

    # Dibuja el histograma
    plt.hist(canal_img.ravel(), bins=256, range=(0, 256), density=True, color='b', alpha=0.7)
    plt.title('Histograma del Canal {}'.format(canal))
    plt.xlabel('Valor de píxel')
    plt.ylabel('Frecuencia')
    plt.show()

    return hist

# Ejemplo de uso
hist = histograma('mario.jpg', 0)
hist1 = histograma('mario.jpg', 1)
hist2 = histograma('mario.jpg', 2)

hist3 = histograma('globo.jpg', 0)
hist4 = histograma('globo.jpg', 1)
hist5 = histograma('globo.jpg', 2)
```
